# Multicollinearity
- occurs when two or more predictor variables in a regression model are highly correlated. This makes it difficult to determine the individual effect of each predictor on the response variable, leading to unreliable and unstable estimates of regression coefficients.

### Key Points:
- Detection: Use Variance Inflation Factor (VIF) or correlation matrices.
- Impact: Inflates variances of estimates, making them sensitive to model changes.
- Solution: Remove or combine correlated predictors, or use dimensionality reduction techniques like PCA.

Standard deviation and variance are fundamental concepts in data science and statistics, helping to describe the spread or dispersion of a dataset.

### Variance
Variance measures how far each data point in a set is from the mean (average) and thus from every other data point. It gives a sense of the data's spread.

#### Formula:
For a population:
\[ \sigma^2 = \frac{\sum (x_i - \mu)^2}{N} \]

For a sample:
\[ s^2 = \frac{\sum (x_i - \bar{x})^2}{n-1} \]

Where:
- \( \sigma^2 \) is the population variance
- \( s^2 \) is the sample variance
- \( x_i \) represents each data point
- \( \mu \) is the population mean
- \( \bar{x} \) is the sample mean
- \( N \) is the number of data points in the population
- \( n \) is the number of data points in the sample

### Standard Deviation
Standard deviation is the square root of the variance, providing a measure of dispersion in the same units as the data, which makes it easier to interpret.

#### Formula:
For a population:
\[ \sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}} \]

For a sample:
\[ s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n-1}} \]

Where:
- \( \sigma \) is the population standard deviation
- \( s \) is the sample standard deviation
- The other variables are as defined above

### Interpretation:
- *Low variance/standard deviation*: Data points are close to the mean, indicating low variability within the dataset.
- *High variance/standard deviation*: Data points are spread out from the mean, indicating high variability within the dataset.

### Usage in Data Science:
1. *Descriptive Statistics*: Summarize and describe the main features of a dataset.
2. *Data Normalization*: Standard deviation is used in techniques like Z-score normalization.
3. *Hypothesis Testing*: Variance and standard deviation are key in various statistical tests.
4. *Machine Learning*: Assess model performance and feature importance.

Understanding and calculating these metrics allow data scientists to gain insights into the data's distribution and variability, aiding in more informed decision-making and analysis.

```python
def variance(data):
  n=len(ages)
  ## mean of the data
  mean=sum(data)/n
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-1)
  return variance

```

```python
variance(ages) # population variance
```

```python
def variance(data,dof=0):
  n=len(ages)
  ## mean of the data
  mean=sum(data)/n
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-dof)
  return variance
```

# Check whether Disstribution is Normal Distribution

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## create a normal distributed data # if the line is stright its means it is normal
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```
